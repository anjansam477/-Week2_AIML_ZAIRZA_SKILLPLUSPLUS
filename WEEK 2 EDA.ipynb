```python
fifa2 = fifa[[column for column in fifa if fifa[column].count() / len(fifa) >= 0.3]]
del fifa2['Name']
del fifa2['ID']
del fifa2['Nationality']
del fifa2['Club']
print("List of dropped columns:", end=" ")
for c in fifa.columns:
    if c not in fifa2.columns:
        print(c, end=", ")
print('\n')
fifa = fifa2
```

```python
fifa_num = fifa.select_dtypes(include = 'number')
fifa_num.head()
```

```python
fifa_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8);
```

```python
fifa_num_corr = fifa_num.corr()['Release Clause'][:-1] # -1 because the latest row is Release Clause 
golden_features_list = fifa_num_corr[abs(fifa_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with Potential:\n{}".format(len(golden_features_list), golden_features_list))
```

```python
for i in range(0, len(fifa_num.columns), 5):
    sns.pairplot(data=fifa_num,
                x_vars=fifa_num.columns[i:i+5],
                y_vars=['Release Clause'])
```

```python
import operator

individual_features_fifa = []
for i in range(0, len(fifa_num.columns) - 1): # -1 because the last column is Release Clause
    tmpDf = fifa_num[[fifa_num.columns[i], 'Release Clause']]
    tmpDf = tmpDf[tmpDf[fifa_num.columns[i]] != 0]
    individual_features_fifa.append(tmpDf)

all_correlations = {feature.columns[0]: feature.corr()['Release Clause'][0] for feature in individual_features_fifa}
all_correlations = sorted(all_correlations.items(), key=operator.itemgetter(1))
for (key, value) in all_correlations:
    print("{:>15}: {:>15}".format(key, value))
```

```python
golden_features_list = [key for key, value in all_correlations if abs(value) >= 0.5]
print("There is {} strongly correlated values with Release Clause:\n{}".format(len(golden_features_list), golden_features_list))
```

```python
corr = fifa_num.drop('Release Clause', axis=1).corr() # We already examined Release Clause correlations
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1, annot=True, annot_kws={"size": 8}, square=True);
```

```python
quantitative_features_list = ['Age','Overall','Potential','Value','Wage','International Reputation','Skill Moves','Height','Weight','Release Clause']
fifa_quantitative_values = fifa[quantitative_features_list]
fifa_quantitative_values.head()
```

```python
features_to_analyse = [x for x in quantitative_features_list if x in golden_features_list]
features_to_analyse.append('Release Clause')
features_to_analyse
```

```python
fig, ax = plt.subplots(round(len(features_to_analyse) / 3), 3, figsize = (18, 12))

for i, ax in enumerate(fig.axes):
    if i < len(features_to_analyse) - 1:
        sns.regplot(x=features_to_analyse[i],y='Release Clause', data=fifa[features_to_analyse], ax=ax)
```

```python
categorical_features = [a for a in quantitative_features_list[:-1] + fifa.columns.tolist() if (a not in quantitative_features_list[:-1]) or (a not in fifa.columns.tolist())]
fifa_categ = fifa[categorical_features]
fifa_categ.head()
```

```python
fifa_not_num = fifa_categ.select_dtypes(include = ['O'])
print('There is {} non numerical features including:\n{}'.format(len(fifa_not_num.columns), fifa_not_num.columns.tolist()))
```

```python
plt.figure(figsize = (10, 6))
ax = sns.boxplot(x='Preferred Foot', y='Release Clause', data=fifa_categ)
plt.setp(ax.artists, alpha=.5, linewidth=2, edgecolor="k")
plt.xticks(rotation=45)
```

```python
plt.figure(figsize = (12, 6))
ax = sns.boxplot(x='Position', y='Release Clause', data=fifa_categ)
plt.setp(ax.artists, alpha=.5, linewidth=2, edgecolor="k")
plt.xticks(rotation=45)
```

```python
fig, axes = plt.subplots(round(len(fifa_not_num.columns) / 3), 3, figsize=(12, 30))

for i, ax in enumerate(fig.axes):
    if i < len(fifa_not_num.columns):
        ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=45)
        sns.countplot(x=fifa_not_num.columns[i], alpha=0.7, data=fifa_not_num, ax=ax)

fig.tight_layout()
```
